# Missing Value Treatment

```python
# Since the number of missing values in these columns are not significant so we impute the missing values with median
df2['AMT_ANNUITY'].fillna(df2['AMT_ANNUITY'].median(),inplace=True)
df2['AMT_GOODS_PRICE'].fillna(df2['AMT_GOODS_PRICE'].median(),inplace=True)
df2['EXT_SOURCE_2'].fillna(df2['EXT_SOURCE_2'].median(),inplace=True)
df2['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(df2['OBS_30_CNT_SOCIAL_CIRCLE'].median(),inplace=True)
df2['OBS_60_CNT_SOCIAL_CIRCLE'].fillna(df2['OBS_60_CNT_SOCIAL_CIRCLE'].median(),inplace=True)
df2['DEF_30_CNT_SOCIAL_CIRCLE'].fillna(df2['DEF_30_CNT_SOCIAL_CIRCLE'].median(),inplace=True)
df2['DEF_60_CNT_SOCIAL_CIRCLE'].fillna(df2['DEF_60_CNT_SOCIAL_CIRCLE'].median(),inplace=True)
df2['DAYS_LAST_PHONE_CHANGE'].fillna(df2['DAYS_LAST_PHONE_CHANGE'].median(),inplace=True)
```

#### Now there are some columns which have a greater proportion of missing values so impute missing values by random sample taken from non null values

```python
def impute_nan(df,variable):
    random_sample = df[variable].dropna().sample(df[variable].isnull().sum(),random_state=0)
    random_sample.index=df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable]=random_sample
```

```python
def one_hot_top_x(data,NAME_INCOME_TYPE,top_x_labels):
    
    
    for label in top_x_labels:
        data[NAME_INCOME_TYPE+"_"+label]=np.where(data[NAME_INCOME_TYPE]==label,1,0)
        
        
one_hot_top_x(data,"NAME_INCOME_TYPE",top4)
data.head()
```
